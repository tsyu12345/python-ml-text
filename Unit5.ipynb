## 5.1:極限と微分

### 5.1.1 極限とは
* 極限：変数の値を特定の値に近づけたとき、その関数の値が限りなくある値に近づくことを指す。
* $例：x = 2のときy = 5, x = 1のときy = 2, ...., x = 0.01のときy = 1.0001$のように$x \rightarrow 0でy \rightarrow 1(xを0に近づけるとyは1に近づく)$のことを以下のように示す。
$$\lim_{x \rightarrow 0} y = 1$$

### 5.1.2 微分とは

関数$y = f(x)$にて、$x$の微小な変化量を$\Delta x$、このときの$y$は以下のように変化する
$$y = f(x + \Delta x)$$
このときの$y$の変化量は$\Delta y$と示す。
$$\Delta y = f(x + \Delta x) - f(x) ※（変化後の値）-（変化前の値）$$

以下のとき、$\Delta x$, $\Delta y$の割合は下記のとおり。
$$\frac{\Delta y}{\Delta x} = \frac{f(x + \Delta x)-f(x)}{\Delta x}$$

* 微分 $\Rightarrow$ 「導関数を得ること」。
* 導関数：$\Delta x \rightarrow 0$の時に得られる新たな関数$f'(x)$
$$f'(x) = \lim_{x \rightarrow 0} \frac{f(x + \Delta x) - f(x)}{\Delta x}$$
* また、微分の記号$dx$を用いて下記の様に示すこともある。
$$f'(x) = \frac{df(x)}{dx} = \frac{d}{dx}f(x)$$
* 常微分：以上のように場合で、関数における変数が1個（1変数関数）の微分を特に”常微分”と呼ぶ。

* 勾配$f'(a)$（こうばい）:$\Delta x$に対する$\Delta y$の割合（変化の割合）$\Rightarrow$ ある点$(a, f'(a))$における接線の傾き
  * 関数$f(x)$上のある点$(a, f(a))$に対応する勾配は$f'(a)$
  * このような関数上のある点における接線の傾き、すなわち勾配は”局所的勾配”と呼ばれる。
  * この接線の式は、$y = f'(a)x + f(a) - f'(a)a$

### 5.1.3 微分の公式

高校数学で学習するような公式

* $\frac{d}{dx}x^r = rx^{r-1}$
* $\frac{d}{dx}(f(x) + g(x)) = f'(x) + g'(x)$
* $\frac{d}{dx}(f(x)g(x)) = f'(x)g(x) + f(x)g'(x)$

### 5.1.4 接線の描画
* $f(x) = 3x^2 + 4x -5$

```python
import numpy as np 
from matplotlib import pyplot as plt
```

```python
def function(x):
    return 3*x**2 + 4*x - 5

def differential_function(x):
    return 6*x + 4

def ready_to_plot(x, y, label:str="") -> None:
    
    plt.plot(x, y, label=label)
    plt.legend()
    plt.xlabel("x")
    plt.ylabel("y")
    plt.grid(True)
    

x = np.linspace(-3, 3)
y = function(x)
a:int = 1

tangent = differential_function(a) * x + function(a) - differential_function(a) * a

ready_to_plot(x, y, label="original function")
ready_to_plot(x, tangent, label="tangent")

plt.show()


```

### 5.1.5 演習
* $f(x) = -2x^2 + x + 3$の$x = 1$における接線

```python
def function(x):
    return -2*x**2  + x + 3

def differential_function(x):
    return -4*x + 1

x_range = np.linspace(-3, 3)
y = function(x_range)
a = 1

tangent = differential_function(a) * x_range + function(a) - differential_function(a) * a

ready_to_plot(x_range, y, label="original function")
ready_to_plot(x_range, tangent, label="tangent")

plt.show()
```

## 5.2 連鎖律（chain rule）

### 5.2.1 合成関数とは
* $y = f(u), u = g(x)$つまり、$y = f(g(x))$という関数を合成関数と呼ぶ。
* 例：$y = (x^2 + 1)^3$として、$g(x) = x^2 + 1, f(g(x)) = g(x)^3$と見ることができる

### 5.2.2 連鎖律とは
合成関数の微分は、それを構成する各関数の微分形（導関数）の積の形で示せる。
これを、”連鎖律”と呼ぶ。
* 連鎖律：合成関数の微分のこと
$$\frac{dy}{dx} = \frac{dy}{du}\frac{du}{dx}$$
* 例:$f(x) = (x^3 + 2x^2 + 3x + 4)^3$の微分
* $ u = x^3 + 2x^2 + 3x + 4 \\
y = u^3 \\ 
として見ることができ、連鎖律を用いて \\ 
\frac{dy}{dx} = \frac{dy}{du}\frac{du}{dx} \\ 
= 3u^2(x^2 + 4x + 3) \\
= 3(x^3 + 2x^2 + 3x + 4)^2(3x^2 + 4x + 3) $

### 5.2.4 演習
* $f(x) = (x^2 + 4x + 1)^4$の微分

$y = u^4$として、$u = x^2 + 4x + 1$とすると,
$\frac{dy}{dx} = 4u^3(2x + 4) = 8u^3(x+2)$\\
$ = 8(x^2 + 4x + 1)^3(x + 2)$

## 5.3 偏微分
偏微分を用いて、あるパラメータの微小な変化が全体に与える影響の大きさを求めることが可能。
### 5.3.1 偏微分とは
* 偏微分：多変数関数を1つの変数に対して微分すること。
  * したがって他の変数は変化しない。$\Rightarrow$ 定数

* 2変数関数$f(x, y)$の偏微分は、
$$\frac{\partial}{\partial x}f(x, y) = \lim_{\Delta x \rightarrow 0}\frac{f(x + \Delta x, y) - f(x, y)}{\Delta x}$$
* この場合、yは変化しない。ので定数扱い。

### 5.3.2 偏微分の例
* $f(x, y) = 3x^2 + 4xy + 5y^3 $の偏微分
  * $x$で偏微分するとき$y$は定数扱いなので、$\frac{\partial}{\partial x}f(x, y) = 6x + 4y$
  * これを偏導関数と呼ぶ。
  * $yの値$を固定した際の、$\Delta x$に対する$f(x, y)$の変化割合になる。
  * $y$で偏微分したときは、同様の手順で、$\frac{\partial}{\partial y}f(x, y) = 4x^2 + 15y^2$

### 5.3.3 演習
* $f(x, y) = 2x^3 + 4x^2y + xy^2 - 4y^2$の偏微分
  * $x$での偏微分
    * $\frac{\partial}{\partial x}f(x, y) = 6x^2 + 8xy + y^2$
  * $y$での偏微分
    * $\frac{\partial}{\partial y}f(x, y) = 4x^2 + 2xy - 8y$

##5.4 全微分
### 5.4.1 全微分とは
* 全微分：多変数関数を全ての変数に対して微分すること。
  * 人工知能では、多くの変数がある多変数関数を扱うので、得られる結果の微小な変化量を得るために全微分を用いる。
### 5.4.3 全微分の例
* $f(x, y) = 3x^2 + 4xy + 5y^3$の全微分
  * それぞれの偏微分を出す。
    * $\frac{\partial}{\partial x}f(x, y) = 6x + 4y$
    * $\frac{\partial}{\partial y}f(x, y) = 4x + 15y$
  
  * 全微分の式$dz = \frac{\partial z}{\partial x}dx + \frac{\partial z}{\partial y}dy$に従い、各偏微分の結果の和をとる。
    * $dz = (6x+4y)dx + (4x+15y^2)dy$

### 5.4.4 演習
* $f(x, y) = 2x^3 + 4x^2y + xy^2 - 4y^2$の全微分
  *それぞれの偏微分は以下の通り。
    * $\frac{\partial}{\partial x}f(x, y) = 6x^2 + 8xy + y^2$
    * $\frac{\partial}{\partial y}f(x, y) = 4x^2 + 2xy - 8y$
  * より、全微分は,
    * $(6x^2 + 8xy + y^2)dx + (4x^2 + 2xy - 8y)dy$

## 5.5 多変数合成関数の連鎖率（合成関数の微分２）
### 5.5.1 多変数合成関数の微分①
*  以下３つの関数を用意する。
  * $z = f(u, v)$
  * $u = g(x)$
  * $v = h(x)$
* 全微分の式より、
  * $dz = \frac{\partial z}{\partial u}du + \frac{\partial z}{\partial v}dv$
  * この式を$dx$で割ることにより「合成関数$z$の$x$による微分」を求めることができる。
  * $\frac{dz}{dx} = \frac{\partial z}{\partial u}\frac{du}{dx} + \frac{\partial z}{\partial v}\frac{dv}{dx}$

### 5.5.2 多変数合成関数の微分②
* 以下の関数を考える。
  * $z = f(u, v)$
  * $u = g(x,y)$
  * $v = h(x,y)$
* $zのxに対する変化割合、zのyに対する変化割合は次式で示される。$
  * $\frac{\partial z}{\partial x} = \frac{\partial z}{\partial u}\frac{\partial u}{\partial x} + \frac{\partial z}{\partial v}\frac{\partial v}{\partial x}$
  * $\frac{\partial z}{\partial y} = \frac{\partial z}{\partial u}\frac{\partial u}{\partial y} + \frac{\partial z}{\partial v}\frac{\partial v}{\partial y}$

### 5.5.3 多変数合成関数の微分例
* 以下の合成関数を$x$で微分するとき
  * $z = u^3 + 3v^2$
  * $u = 2x^2 + 3x + 4$
  * $v = x^2 + 5$
* 全微分の式から、
  * $\frac{dz}{dx} = 3u^2(4x+3) + 6v(2x) = 3(2x^2 + 3x + 4)^2 + (4x + 3) + 12x(x^2 + 5)$

* 多変数合成関数でも連鎖率による微分ができる。

### 5.5.4 演習
* 次の合成関数を微分しろ。
  * $z = 2u^3 + uv^2 + 4v$
  * $u = x^2 + 3x$
  * $v = x + 7$
* 【解答】
  * $\frac{du}{dx} = 2x + 3$
  * $\frac{dv}{dx} = 1$
  * より,全微分の式から
    * $\frac{dz}{dx} = (6u^2 + v^2)(2x + 3) + (2uv + 4) = (6(x^2 + 3x)^2 + (x+7)^2)(2x + 3) + 2(x^2 + 3x)(x + 7)$

### 5.6.6 シグモイド関数
* $y = \frac{1}{1 + e^{-x}} = \frac{1}{1 + \exp(-x)}$
* その導関数は$(1-y)y※導出はP174$

```python
def sigmoid(x):
    return 1/(1 + np.exp(-x))

def sigmoid_differential(x):
    y = sigmoid(x)
    return y * (1 - y)


x = np.linspace(-5, 5)
y = sigmoid(x)
y_differential = sigmoid_differential(x)

ready_to_plot(x, y, label="sigmoid")
ready_to_plot(x, y_differential, label="sigmoid differential")
plt.show()
```

* シグモイド関数は、入力値を$0~1$の範囲に変形する役割をもつ。※青グラフ参照。
* 人工知能の「活性化関数」として使われる。

## 5.7 最急降下法
### 5.7.1 最急降下法とは
* 勾配法：関数の微分値をもとに、その関数の最小値の探索をするアルゴリズムの総称。
* 最急降下法：勾配法のうち、最も勾配（ある点における接線の傾き）が急な箇所を目指すようにして最小値の探索を行う方法。
* 変数が$n$個の多変数関数$f(\vec{x}) = f(x_1, x_2, ...., x_n)$の最小値を探索する。
  * $\vec{x}の初期値は適当で、x_i(i = 1, 2, ...n)$を学習係数（定数）$\eta$を用いて更新する。
  * $$x_i = x_{i(prev)}-\eta \frac{\partial f(\vec{x})}{\partial x_{i(prev)}}$$
  * 上記式を全要素に対して、$f(\vec{x})$の全要素に変化がなくなるまで繰り返す。



### 5.7.2 最急降下法の実装
* $f(x) = x^2 -2x$の最小値を最急降下法で求める。
  * $f'(x) = \frac{df(x)}{dx} = 2x-2$

```python
from __future__ import annotations

    
class SteepestDescent():
    
    def __init__(self, eta:float, x:int|float, renew_count:int):
        self.renew_count = renew_count
        self.eta = eta
        self.x = x
        self.function = function(x)
        self.diff_function = function_differential(x)
    
    def function(self, x:int|float) -> float:
        return x**2 - 2*x

    def function_differential(self, x) -> float:
        return 2*x - 2
    
    
    def plot_graph(self) -> None:
        record_x:list[float] = []
        record_y:list[float] = []
        for i in range(self.renew_count):
            y = self.function(self.x)
            record_x.append(self.x)
            record_y.append(y)
            
            self.x -= self.eta * self.function_differential(self.x)
            
        x_f = np.linspace(-2, 4)
        y_f = self.function(x_f)
        
        plt.plot(x_f, y_f, linestyle="dashed")
        plt.scatter(record_x, record_y)
        
        plt.grid()
        plt.show()
        
t = SteepestDescent(0.11, 4.0, 20)
t.plot_graph()
```

### 5.7.3 局所的な最小値

* このような場合、極小値は複数存在するケースがある。「全体の最小値（本当の意味での最小値）」と「局所的な最小値（一時的に勾配が０になる箇所）」の２つがある。
* このときの値を「局所最適解、停留点」と言ったりする。

* $f(x) = x^4 + 2x^3 -3x^2 -2x$とし、
* $\frac{df(x)}{dx} = 4x^3 + 6x^2 -6x -2$
*この関数の最小値を最急降下法で求めると以下のようになる。

```python
def function(x:float):
    return x**4 - 2*x**3 + 3*x**2 - 2*x

def function_differential(x:float):
    return 4*x**3 - 6*x**2 + 6*x - 2

SteepestDescent(0.11, 4.0, 20).plot_graph()
```
